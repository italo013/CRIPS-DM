<h1>Clean Data</h1>

```python
df.head()
```

```python
#Tratando a coluna usd_goal_real, removendo a string e deixando apenas os valores num√©ricos
df['usd_goal_real'] = df['usd_goal_real'].str.lstrip('USD').astype('int64')
#df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD ')).astype('float64')

#Removendo a hora da coluna launched
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

<h1>Construct Data</h1>

```python
df['time_range'] = (df['deadline'] - df['launched']).dt.days
```

```python
df.head()
```

```python
df_invest = pd.get_dummies(raw_df_invest, columns=['backedLocation']).sort_values(by=['ID'])
```

```python
df_invest.head()
```

<h1>Integrate Data</h1>

```python
df_full = df.merge(df_c, how='right', on=['ID'])
```

```python
df_invest = df_invest.groupby(by=['ID']).agg({'age': 'mean', 'backedLocation_BR': 'sum', 'backedLocation_GBK': 'sum', 'backedLocation_US': 'sum'})
```

```python
df_full = df_full.merge(df_invest, how='right', on='ID')
df_full.head()
```

<h1>Format Data</h1>

```python
#Transformando as colunas no tipo data no padr√£o americano
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```
